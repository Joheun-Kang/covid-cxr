## Why Interpretability Important?  <br>

A lot of companies these days use machine learning algorithms for decision making. In the customer perspective, the customer wants to know why that company made the decision for their requests.<br> 

For example, the bank may or may not accept their customer's loan by some features. And, the customers want to know why the bank give them accepted or declined on their requests. <br> 

For another example. Let's say we have a picture of a dog, and we use a machine-learning algorithm to classify the dog's breed or whether that is a dog or not. <br>

Let's say the M.L model made a decision that this dog is a wolf. 
The decision was not what we expected, so we want to know why this decision is made!<br>

In this case, we use SHAP to explain the reasons behind the decisions. <br>

## Shapley Value <br>


In order to explain our result from the model that we used, we definitly want to explain which features and how much the features affect to our result. Most of the time, features are not independent to each others. Because of that reason, the features on the model's final prediction is not trivial. This is why we need the Shapley concept. <br>


For better understading, I will explain using the example in the __Google's " A.I Exaplainability Whitepaper"__<br>

Assume that a company with a set N of three employees {A.B.C}. and the profit outcome of v(N).<br>
We want to distribute the profit among those three people based on how much they contribute to get the profit. We are going to find all the combinations of the three employee that improve the profit compared to the foil = {} (no employees). 

Let's assume <br>

v({}) = 0 <br>
v({A}) = 10 <br>
v({B})= 20 <br>
v({C}) = 30 <br>
v({A, B}) = 60 <br>
v({B, C}) = 70 <br>
v({A, C}) = 90 <br>
v({A, B, C}) = 100 <br>

Important to note here is that when we add B to the v({A}), B gives us 50 more profit than A alone. However, this does not mean that B always increase profit by 50. As we can see from the above, if we add B to the v({C}), we get 40 more profit. <br>

To account for this, the Shapley value considers all of the permutations such as <br>

1. {}{A}{A, B}{A, B, C}<br>
2. {}{A}{A, C}{A, B, C}<br>
3. {}{B}{A, B}{A, B, C}<br>
4. {}{B}{B, C}{A, B, C}<br>
5. {}{C}{B, C}{A, B, C}<br>
6. {}{C}{A, C}{A, B, C}<br>

for each case, we compute a credit <br>

A: v({A})v({}) = 100 = 10 <br>
B: v({A,B})v({A}) = 6010 = 50 <br>
C: v({A,B,C})  v({A,B}) = 100  60 = 40 <br>

so the result, <br>
1. {}{A}{A, B}{A, B, C}||A = 10, B = 50, C = 40
2. {}{A}{A, C}{A, B, C}||A = 10, B = 10, C = 80
3. {}{B}{A, B}{A, B, C}||A = 40, B = 20, C = 40
4. {}{B}{B, C}{A, B, C}||A = 30, B = 20, C = 50
5. {}{C}{B, C}{A, B, C}||A = 30, B = 40, C = 30
6. {}{C}{A, C}{A, B, C}||A = 60, B = 10, C = 30 <br>

__Shapley values: A_{avg} = 30 , B_{avg} = 25 , C_{avg} = 45__ <br>


From the Shapley values, we know how much those feature contribute for our result.<br>

We will see why the shapley values are important, and how we use the those values from the following part. 











## SHAP (SHapley Additive exPlanation) <br>
SHAP is one of the frameworks enables any Machine Learning algorithm to become "explain" by providing intuitive and visualizations for showing which features are more relevant for a certain prediction and also for the model as a whole.<br>

In this section, I will explain why it is good to use SHAP, especially for the deep learning models. If you want the math behind the SHAP, please use the paper as a reference. 

Deep learning models have revolutionized the machine learning field. Most of the deep learning models give us noticeably accurate results (predictions) than typical machine learning models. However, since deep learning models complexity, it is more difficult to explain the reasons than typical machine learning models.

As we learned in class, you can think of it by substituting a gradient concept, but there are limits to explaining the results of the models. However, using the built-in SHAP, we can visualize the predictions (results) as well as the result of each layer that helps our results.


The image below shows what a prediction was made for a particular layer in the VGG16 model. Shapley value has been used for each layer to described above in each of these layers. The blue color with a Shapley value less than 0 is the part that does not affect (negative effect) to the prediction, and the red pixels are the pixels that gave positive effects for the prediction (label).




Also, usually. If the results are different from our expectations (or certain measures), it is difficult to figure out where the issue came from. We judge the overall ability of this model by looking at the results only. However, if we use SHAP, we can figure out where the issues occurred.





The image shown above is the result of the model's poor prediction. But we don't know why the model came to this conclusion only by looking at the results. However, if we can check the image per layer, we know which layers put incorrect weights for the features that affect incorrect prediction (i.e pen scratches in the X-ray). 

As we have seen from the above examples, we can explain the reasons behind the model's prediction. <br>
